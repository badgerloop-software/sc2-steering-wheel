```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the file
data = np.loadtxt('BSRlogo_array.txt')

# Display the shape of the array
print(data.shape)

# Visualize the data
plt.imshow(data)
plt.title('BSR Logo')
plt.show()
```

```python
height, width = data.shape
print(height, width)
```

```python
from PIL import Image
```

```python
data = data.flatten()
data.shape
```

```python
####Format used is RRRGGGBB

data_int = data.astype(np.uint8)

r_recon = np.uint8(((data_int >> 5) & 0x07) * (255.0 / 7.0)).reshape((height, width))
g_recon = np.uint8(((data_int >> 2) & 0x07) * (255.0 / 7.0)).reshape((height, width))
b_recon = np.uint8((data_int & 0x03) * (255.0 / 3.0)).reshape((height, width))

recon_array = np.zeros((height, width, 3), dtype=np.uint8)
recon_array[:,:,0] = r_recon
recon_array[:,:,1] = g_recon
recon_array[:,:,2] = b_recon

Image.fromarray(recon_array)
```
